Null hyposthesis=There is no difference in diameters of cutlets between two units.
Alternate hypothesis=There is difference between diameters of cutlets of two units.

```python
# Two Tail test
stats.ttest_ind(cutlet['Unit A'], cutlet['Unit B'])
```

Here we can see pvalue > 0.05. We can say that there is no difference in diameters of cutlets between two units.
