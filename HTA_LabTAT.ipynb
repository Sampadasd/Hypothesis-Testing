Null Hypothesis=All samples TAT population means are same.
Alternate Hypothesis=Atleast one sample TAT population mean is different.

```python
#Anova ftest statistics
stats.f_oneway(labtat.iloc[:,0], labtat.iloc[:,1],labtat.iloc[:,2],labtat.iloc[:,3])
```

Here we can see pvalue < 0.05. We can say that atleast one sample TAT population mean is different.
